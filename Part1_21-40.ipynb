```python
# Q-21 Write a Python program that determines whether a given number (accepted from the user) is even or odd, and prints an appropriate message to the user.

num = int(input("Enter any Number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")    
```

```python
# Q-22 Write a Python program to count the number 4 in a given list.

l1 = input("Enter elements seperated by space:").split()
print("List: ",l1)

count=0

for i in range(len(l1)):
    if l1[i] == "4":
        count = count+ 1

print(count)   


l2 = input("Enter elements seperated by space: ").split()
print("Count of 4: ", l2.count("4"))
```
