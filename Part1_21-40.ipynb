```python
# Q-21 Write a Python program that determines whether a given number (accepted from the user) is even or odd, and prints an appropriate message to the user.

num = int(input("Enter any Number: "))

if num % 2 == 0:
    print("Even Number")
else:
    print("Odd Number")    
```

```python
# Q-22 Write a Python program to count the number 4 in a given list.

l1 = input("Enter elements seperated by space:").split()
print("List: ",l1)

count=0

for i in range(len(l1)):
    if l1[i] == "4":
        count = count+ 1

print(count)   


l2 = input("Enter elements seperated by space: ").split()
print("Count of 4: ", l2.count("4"))
```

```python
# Q-23 Write a Python program to get n (non-negative integer) copies of the first 2 characters of a given string. Return n copies of the whole string if the length is less than 2.

str1 = input("Enter any String :")
n = abs(int(input("Enter number to print copies of string: ")))

if len(str1)<2:
    for i in range(n):
        print(str1)
else:
    for i in range(n):
        print(str1[:2])        
```

```python
# Q-24 Write a Python program to test whether a passed letter is a vowel or not.

letter = input("Enter a letter: ")
vowels = ['a','e','i','o','u']
if letter not in vowels :
    print("It is not a Vowel")
else:
    print("It is a Vowel.")    
```
