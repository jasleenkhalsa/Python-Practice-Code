```python
# Q4 - Write a Python program that calculates the area of a circle based on the radius entered by the user

import math

def area(radius):
    return math.pi * radius ** 2

radius = float(input("Enter the Radius of the Circle: "))

a = area(radius)

print(f"The area of the circle is: {a}")
```

```python
# Q-14 Write a Python program to calculate the number of days between two dates.

from datetime import datetime

date1 = input("Enter first date (YYYY-MM-DD): ")
date2 = input("Enter second date (YYYY-MM-DD): ")

d1 = datetime.strptime(date1, "%Y-%m-%d")
d2 = datetime.strptime(date2, "%Y-%m-%d")

difference = abs((d2-d1).days)

print(difference)
```

```python
# Q-16 Write a Python program to calculate the difference between a given number and 17. If the number is greater than 17, return twice the absolute difference.

num = int(input("Enter a Number :"))

if num <= 17:
    print(f"The Difference is {17-num}")
else:
    print(f"The Twice of Difference is {(num-17)*2}")    
```

```python
# Q-17 Write a Python program to test whether a number is within 100 of 1000 or 2000.

num = int(input("Enter a Number :"))

if (999<num and num<1101):
    print("The Number is within 100 of 1000.")
elif (1999<num and num<2101):
    print("The Number is within 100 of 2000.")
else:
    print("The Number is neither within 100 of 1000 nor 2000")
```

```python
# Q-18 Write a Python program to calculate the sum of three given numbers. If the values are equal, return three times their sum.

num1 = int(input("Enter the First number: "))
num2 = int(input("Enter the Second number: "))
num3 = int(input("Enter the Third number: "))

if (num1==num2 and num2==num3):
    print("The Thrice of the Sum is : ",(num1+num2+num3)*3)
else:
    print("The Sum is :",(num1+num2+num3))
```

```python
# Q-19 Write a Python program to get a newly-generated string from a given string where "Is" has been added to the front. Return the string unchanged if the given string already begins with "Is".

string = input("Enter a string: ")

if string[:2] == "Is":
    print(string)
else:
    print("Is " + string)
```

```python
# Q-20 Write a Python program that returns a string that is n (non-negative integer) copies of a given string.

string = input("Enter the String")
n = int(input("Enter the number of times the copies should be generated :"))

for i in range(n):
    print(string)
```
